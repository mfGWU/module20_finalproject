```python
from path import Path
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Files to load 
colorado_offense_to_load_da = "file_encoded.csv"
```

```python
# Read the DenverCrime file and store it in a Pandas DataFrame.
colorado_offense_to_load_da = pd.read_csv(colorado_offense_to_load_da)
colorado_offense_to_load_da
```

# Separate the Features (X) from the Target (y)

```python
y = colorado_offense_to_load_da["MONTH_NUM"]
X = colorado_offense_to_load_da.drop(columns="MONTH_NUM")
```

# Split our data into training and testing

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1, 
                                                    stratify=y)
X_train.shape
```

# Create a Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs',
                                max_iter=200,
                                random_state=1)
```

# Fit (train) or model using the training data

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
results = pd.DataFrame({"Prediction": y_pred, "Actual": y_test}).reset_index(drop=True)
results.head(20)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```
