# Dealing with Categorical Data in ML

### Dataset Information

The file NIBRS_incident.csv, contains State of Colorado crime incidents information from 2016 till 2019, there are a total of 54,930 records. The file NIBRS_Offense.csv, contains offense record State of Colorado associated to each crime incidentf rom 2016 till 2019, there are a total of 54,930 records. The features listed below do not represent the complete list from each file and listed features and target are going to be pulled into one dataframe to run predictive model.

NIBRS_incident Dataset: 
- DATA_YEAR INT: The year the incident/offense took place.
- AGENCY_ID INT: Agency responding to the incident.
- INCIDENT_ID INT: The automated incident number.
- NIBRS_MONTH_ID INT: Some ID for the month.
- CARGO_THEFT_FLAG VARCHAR:	Any item taken during the incident (Y/N).
- SUBMISSION_DATE DATE: Submission date.
- INCIDENT_DATE INT: The date the incident/offense took place.
- INCIDENT_Month VARCHAR: Month the incident/offense took place.
- INCIDENT_DAY INT: Day the incident/offense took place.
- REPORT_DATE_FLAG VARCHAR: Report flagged.
- INCIDENT_HOUR INT: Hour the incident/offense took place.
- DATA_HOME VARCHAR: State the incident was reported from (C = Colorado).
- ORIG_FORMAT VARCHAR: Original Format.
- DID INT:

NIBRS_Offense Dataset:
- DATA_YEAR INT:Year date when offense was entered.
- OFFENSE_ID INT: The automated offense number.
- INCIDENT_ID INT: Foreign Key
- OFFENSE_TYPE_ID VARCHAR: Offense commiteed by the perpetrator (Burglary/Breaking & Entering,Robbery)
- ATTEMPT_COMPLETE_FLAG VARCHAR: Field hat holds F = Force and N = Nonforce.
- LOCATION_ID VARCHAR: Where the perpetrator committed the offense (Residence/Home,Commercial/Office Building, Government/Public Building)
- METHOD_ENTRY_CODE VARCHAR: How did the perpetrator entered the premise (F, N).


```python
# initial imports
import pandas as pd
from path import Path
```

```python
# Files to load 
colorado_offense_to_load_da = "Resources/NIBRS_Offense_16_19.csv"
colorado_incidents_to_load_da = "Resources/Nibrs_incidents_16_19.csv"
```

```python
# Read the DenverCrime file and store it in a Pandas DataFrame.
colorado_incidents_da = pd.read_csv(colorado_incidents_to_load_da)
colorado_incidents_da
```

```python
# Read the NIBRS_Offense_16_19.csv file and store it in a Pandas DataFrame.
colorado_offense_da = pd.read_csv(colorado_offense_to_load_da)
colorado_offense_da
```

```python
colorado_incidents_da_df=pd.DataFrame(data=colorado_incidents_da)
colorado_incidents_da_df
```

```python
colorado_offense_da_df=pd.DataFrame(data=colorado_offense_da)
colorado_offense_da_df
```

```python
# Merge two Pandas DataFrames on multiple columns in Python

#pd.concat([colorado_incidents_da_df,colorado_offense_da_df], ignore_index=True)
merged_df = pd.merge(colorado_incidents_da_df, colorado_offense_da_df, how="inner",
    on=['INCIDENT_ID', 'INCIDENT_ID'])
merged_df
```

```python
# Select multiple columns to create new dataframe
incident_offense_df = merged_df[['INCIDENT_ID','OFFENSE_ID','﻿DATA_YEAR_x','INCIDENT_Month','INCIDENT_DAY','INCIDENT_HOUR','OFFENSE_TYPE_ID','LOCATION_ID']]
incident_offense_df
```

```python
# Custom Encode the months
months_num = {
   "January": 1,
   "February": 2,
   "March": 3,
   "April": 4,
   "May": 5,
   "June": 6,
   "July": 7,
   "August": 8,
   "September": 9,
   "October": 10,
   "November": 11,
   "December": 12,
}
```

```python
# A lambda function is applied to the month column to perform the actual conversion
incident_offense_df["MONTH_NUM"] = incident_offense_df["INCIDENT_Month"].apply(lambda x: months_num[x])
incident_offense_df
```

```python
# Drop the "INCIDENT_Month" column after converting the months to number
incident_offense_df = incident_offense_df.drop(["INCIDENT_Month"], axis=1)
incident_offense_df.head()
```

```python
# Binary encoding using Pandas (multiple columns)
incident_offense_df_encoded = pd.get_dummies(incident_offense_df, columns=["OFFENSE_TYPE_ID", "LOCATION_ID"])
incident_offense_df_encoded
```

# Predicting Crime

Implementing Logistic Regression model with incident_offense_df to predict when a crime will take place by targeting "MONTH_NUM". It will follow the following pattern: instantiate a model, train it, create predictions, then validate the model.

## Separate the Features (X) from the Target (y)

```python
y = incident_offense_df_encoded["MONTH_NUM"]
X = incident_offense_df_encoded.drop(columns="MONTH_NUM")
```

## Split our data into training and testing

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1, 
                                                    stratify=y)
X_train.shape
```

## Create a Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs',
                                max_iter=200,
                                random_state=1)
```

## Fit (train) or model using the training data

```python
classifier.fit(X_train, y_train)
```
