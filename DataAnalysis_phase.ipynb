## Dealing with Categorical Data in ML

### Dataset Information

The file NIBRS_incident.csv, contains State of Colorado crime incidents information from 2016 till 2019, there are a total of 54,930 records. The file NIBRS_Offense.csv, contains offense record State of Colorado associated to each crime incidentf rom 2016 till 2019, there are a total of 54,930 records. The features listed below do not represent the complete list from each file and listed features and target are going to be pulled into one dataframe to run predictive model.

NIBRS_incident Dataset: 
- DATA_YEAR INT: The year the incident/offense took place.
- INCIDENT_ID INT: The automated incident number.
- INCIDENT_DATE INT: The date the incident/offense took place.
- INCIDENT_Month VARCHAR: Month the incident/offense took place.
- INCIDENT_DAY INT: Day the incident/offense took place.
- INCIDENT_HOUR INT: Hour the incident/offense took place.

NIBRS_Offense Dataset:
- OFFENSE_ID INT: The automated offense number.
- INCIDENT_ID: Foreign Key
- OFFENSE_TYPE_ID VARCHAR: Offense commiteed by the perpetrator (Burglary/Breaking & Entering,Robbery)
- LOCATION_ID VARCHAR: Where the perpetrator committed the offense (Residence/Home,
- Commercial/Office Building, Government/Public Building)
